### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "C:/Users/golfz/Documents/GitHub/PyBer_Analysis/Resources/city_data.csv"
ride_data_to_load = "C:/Users/golfz/Documents/GitHub/PyBer_Analysis/Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

```python
pyber_data_df
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides
```

```python
# 2. Get the total drivers for each city type
total_drivers = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers
```

```python
#  3. Get the total amount of fares for each city type
total_fares = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares
```

```python
#  4. Get the average fare per ride for each city type. 
urban_avg_fare = total_fares["Urban"]/total_rides["Urban"]
suburban_avg_fare = total_fares["Suburban"]/total_rides["Suburban"]
rural_avg_fare = total_fares["Rural"]/total_rides["Rural"]
print(urban_avg_fare)
print(suburban_avg_fare)
print(rural_avg_fare)
```

```python
# 5. Get the average fare per driver for each city type. 
urban_avg_fare_drivers = total_fares["Urban"]/total_drivers["Urban"]
suburban_avg_fare_drivers = total_fares["Suburban"]/total_drivers["Suburban"]
rural_avg_fare_drivers = total_fares["Rural"]/total_drivers["Rural"]
print(urban_avg_fare_drivers)
print(suburban_avg_fare_drivers)
print(rural_avg_fare_drivers)
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({'Total Rides':total_rides, 'Total Drivers':total_drivers, 'Total Fares':total_fares, 'Average Fare per Ride':[rural_avg_fare, suburban_avg_fare, urban_avg_fare], 'Average Fare per Driver':[rural_avg_fare_drivers, suburban_avg_fare_drivers, urban_avg_fare_drivers]})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df['Total Fares'] = pyber_summary_df['Total Fares'].map("${:,.2f}".format)
pyber_summary_df['Average Fare per Ride'] = pyber_summary_df['Average Fare per Ride'].map("${:,.2f}".format)
pyber_summary_df['Average Fare per Driver'] = pyber_summary_df['Average Fare per Driver'].map("${:,.2f}".format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Read the merged DataFrame
pyber_summary_df
```

```python
# 9. Use groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
pyber_date_df = pyber_data_df[['type', 'date', 'fare']]
pyber_date_summary_df = pyber_date_df.groupby(["type", "date"]).sum()["fare"]
pyber_date_summary_df
```

```python
# 10. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
pyber_date_summary_df = pyber_date_summary_df.reset_index()
pyber_date_summary_df
```

```python
# 11. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_date_summary_df = pyber_date_summary_df.pivot(index = 'date', columns = 'type', values = 'fare')
pyber_date_summary_df
```

```python
# 12. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pyber_date_summary_df_q1_2019 = pyber_date_summary_df.loc['2019-01-01':'2019-04-29']
pyber_date_summary_df_q1_2019
```

```python
# 13. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 15.
# df.index = pd.to_datetime(df.index)
pyber_date_summary_df_q1_2019.index = pd.to_datetime(pyber_date_summary_df_q1_2019.index)
pyber_date_summary_df_q1_2019
```

```python
# 14. Check that the datatype for the index is datetime using df.info()
pyber_date_summary_df_q1_2019.info()
```

```python
# 15. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pyber_date_summary_df_q1_2019 = pyber_date_summary_df_q1_2019.resample('W').sum()
pyber_date_summary_df_q1_2019
```

```python
# 16. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
import matplotlib.pyplot as plt
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

# Create the plot
fig, ax = plt.subplots(figsize=(16, 8))
pyber_date_summary_df_q1_2019.plot(ax=ax, title = "Total Fare by City Type")
ax.set(ylabel = "Fare ($USD)", xlabel = "Date(Month)")

# Save the figure.
plt.savefig("C:/Users/golfz/Documents/GitHub/PyBer_Analysis/analysis/Pyber_fare_summary.png")
```
